```python
import json
from werkzeug.utils import secure_filename

with open('archives.json', 'r') as f:
    archive_dict = json.load(f)

for key, val in archive_dict.items():
    filename = f"{secure_filename(val['title'])}.json"
    if 'http' in val['thumbnail']:
        val['thumbnail'] = 'assets/thumbnails/' + val['id'] + '.webp'
    val['filepath'] = f'archives/{val['id']}.mp3'
    val['filename'] = val['id'] + '.mp3'
    val['id'] = filename.replace('.json','')
    if 'Reading For Now' in val['title']:
        val['show'] = 'r'
    elif 'Cuts In The Fog' in val['title']:
        val['show'] = 'c'
    elif 'Afternoon Breakfast' in val['title'] or 'Moonlighting' in val['title']:
        val['show'] = 'a'
    with open(f'data/{filename}', 'w') as f:
        json.dump(val, f)
```

```python
import json
import os
import random 
import string

archive_dict = {}
missing_files = []
archive_data = os.listdir('data')
for archive_file in archive_data:
    if archive_file.endswith('.json'):
        with open(f'data/{archive_file}', 'r') as f:
            data = json.load(f)
            archive_id = data['id']
            archive_dict[archive_id] = data

new_archive = {}
for key, val in archive_dict.items():
    stringy = ''.join(random.choices(string.ascii_letters + string.digits, k=16))
    new_archive[stringy] = val
    new_archive[stringy]['id'] = stringy
    with open(f'data/{stringy}.json', 'w') as f:
        json.dump(new_archive[stringy], f)

new_archive
```
