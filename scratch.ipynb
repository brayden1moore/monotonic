```python
import requests
from bs4 import BeautifulSoup

def monotonic_live_link():
    url = 'https://monotonic.studio/live'
    webpage = requests.get(url).text
    soup = BeautifulSoup(webpage, 'html.parser')
    iframes = soup.find_all('iframe')
    for i in iframes:
        if 'meshcast' in i['src']:
            return i['src']

monotonic_live_link()
```
