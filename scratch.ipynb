```python
import json
from werkzeug.utils import secure_filename

with open('archives.json', 'r') as f:
    archive_dict = json.load(f)

for key, val in archive_dict.items():
    filename = f"{secure_filename(val['title'])}.json"
    if 'http' in val['thumbnail']:
        val['thumbnail'] = 'assets/thumbnails/' + val['id'] + '.webp'
    val['filepath'] = f'archives/{val['id']}.mp3'
    val['id'] = filename.replace('.json','')
    if 'Reading For Now' in val['title']:
        val['show'] = 'r'
    elif 'Cuts In The Fog' in val['title']:
        val['show'] = 'c'
    elif 'Afternoon Breakfast' in val['title'] or 'Moonlighting' in val['title']:
        val['show'] = 'a'
    with open(f'data/{filename}', 'w') as f:
        json.dump(val, f)
```
